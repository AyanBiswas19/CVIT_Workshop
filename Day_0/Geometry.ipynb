```python
def distance_from_origin(a,b,c):
    return abs(c)/((a**2+b**2)**0.5)

def distance_between_parallel_lines(a1,b1,c1,a2,b2,c2):
    return abs(c1-c2)/((a1**2+b1**2)**0.5)
```

## Plotting functions using linspace#

```python
#sigmoid function 
X = np.linspace(-10,10,200)
Y = 1/(1 + np.exp(-X))
plt.plot(X,Y)
```

```python
#gaussian function 
X = np.linspace(-10,10,200)
Y = np.exp(-(X**2))
plt.plot(X,Y)
```

```python
#rectified linear unit - ReLU
X = np.linspace(-10,10,200)
Y = np.maximum(0,X)
plt.plot(X,Y)
```

```python
#Leaky ReLU
X = np.linspace(-10,10,200)
Y = np.maximum(0.1*X,X)
plt.plot(X,Y)
```

```python
#Step Function
def step_func(x):
    return 1 if x >= 0 else 0
vsf = np.vectorize(step_func)
X = np.linspace(-10,10,200)
Y = vsf(X)
plt.plot(X,Y)
```

## Plotting shapes using linspace

```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

#Parameters
a=1
b=1
c=0
d=10

xx = np.linspace(-15,15,200)
zz = np.linspace(-15,15,200)
X,Z = np.meshgrid(xx,zz)

# ax + by + cz + d = 0
Y = (-d -a*X -c*Z) / b

plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(X,Y,Z)
```

```python
#make circle
r = 2
X = np.linspace(-r,r,200)
Yp = np.sqrt(r**2 - X**2)
Yn = -Yp 
plt.plot(X,Yp)
plt.plot(X,Yn)
plt.axis('equal')
```

```python
#make elipse 
a = 5
b = 3
X = np.linspace(-a,a,200)
Yp = np.sqrt(1 - X**2/a**2) * b
Yn = -Yp 
plt.plot(X,Yp)
plt.plot(X,Yn)
plt.axis('equal')
```

```python
#make square
a = 5
npoints_per_line = 200

l1_x = np.zeros(npoints_per_line)
l1_y = np.linspace(0,a,npoints_per_line)

l2_x = np.linspace(0,a,npoints_per_line)
l2_y = np.zeros(npoints_per_line)

l3_x = np.zeros(npoints_per_line) + a
l3_y = np.linspace(0,a,npoints_per_line)

l4_x = np.linspace(0,a,npoints_per_line)
l4_y = np.zeros(npoints_per_line) + a

plt.plot(l1_x,l1_y)
plt.plot(l2_x,l2_y)
plt.plot(l3_x,l3_y)
plt.plot(l4_x,l4_y)
plt.axis('equal')
```

```python
#make rectangle
a = 5
b = 10
npoints_per_line = 200

l1_x = np.zeros(npoints_per_line)
l1_y = np.linspace(0,a,npoints_per_line)

l2_x = np.linspace(0,b,npoints_per_line)
l2_y = np.zeros(npoints_per_line)

l3_x = np.zeros(npoints_per_line) + b
l3_y = np.linspace(0,a,npoints_per_line)

l4_x = np.linspace(0,b,npoints_per_line)
l4_y = np.zeros(npoints_per_line) + a

plt.plot(l1_x,l1_y)
plt.plot(l2_x,l2_y)
plt.plot(l3_x,l3_y)
plt.plot(l4_x,l4_y)
plt.axis('equal')
```

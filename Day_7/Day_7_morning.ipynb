```
flag = cv2.imread('/content/drive/My Drive/my_test_images/flag_of_india.JPG', cv2.IMREAD_GRAYSCALE)
cv2_imshow(flag)
print(flag.shape)
print(flag.shape[0] * flag.shape[1])
```

Let's use svd on this

```
u, d, vt = np.linalg.svd(flag, full_matrices = False)
```

```
a = u @ np.diag(d) @ vt
cv2_imshow(a)
```

```
components =  50
flag_compressed = u[:, 0:components] @ np.diag(d[0:components]) @ vt[0:components]
cv2_imshow(np.hstack([flag, flag_compressed]))
h, w = flag.shape
ratio = (h*w)/(h*components + components + components*w)
print(ratio)
```
